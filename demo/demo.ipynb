## Test sledgehammer

Earlier we invoked sledgehammer by sending the literal `sledgehammer` Isar command as a proof **step** (`execute_command(..., 'sledgehammer')`). Running it that way executes inside the document node and blocks the session's other ML channels (e.g. `open_subgoals`) until it finishes.

Here we use the **standalone sledgehammer channel** instead: `client.sledgehammer(session_id, ...)` hits the dedicated `/sledgehammer` endpoint. On the ML side it forks an Isabelle thread, so it does **not** run as a proof step, does not advance/modify the proof state (read-only), and does not block other sessions. It returns the proof-method suggestions (e.g. `by (metis ...)`) which the prover can then apply as the next step.

## Test session parallelism (`parallel_proofs` / `threads`)

Sessions are now created with configurable PIDE parallelism (`repl/src/main/scala/repl/session_manager.scala`):

- `ISABELLE_PARALLEL_PROOFS` (default `2`) — `2` forks nested `have`/`show` proof bodies so **independent `have`s are checked concurrently**; `1` = top-level proofs only; `0` = fully sequential.
- `ISABELLE_SESSION_THREADS` (default `4`) — caps the ML worker threads per session (deliberately not `0` = auto, so concurrent sessions don't oversubscribe cores and OOM the gateway).

Both are read from the **server's** environment at session creation — set them in `.env` at the repo root and restart the server to change them. Pooled sessions keep the options they were created with.

**The experiment:** one lemma with `N_HAVES` independent `have` steps, each proof sleeping `SLEEP_S` seconds inside an ML tactic before closing the goal. Submitted as **one chunk**, PIDE forks the `have` proofs, so wall time ≈ `ceil(N_HAVES / threads) * SLEEP_S`. The same steps submitted **one command at a time** form the sequential baseline (~`N_HAVES * SLEEP_S`), because each small-step waits for the node to finish checking before returning.

````python
session_payload = await client.acquire_session(["Complex_Main"], "HOL")
session_id = session_payload.get("session_id")
lease_id = session_payload.get("lease_id")
session_payload
````

````python
N_HAVES = 4
SLEEP_S = 2

def slow_have(i: int) -> str:
    # Independent goal; the ML tactic sleeps, then closes it with simp.
    # With parallel_proofs=2 these `by` bodies are forked and run concurrently.
    return (
        f'have h{i}: "{i} + 0 = ({i}::nat)"\n'
        f"  by (tactic ‹(OS.Process.sleep (Time.fromSeconds {SLEEP_S}); simp_tac @{{context}} 1)›)"
    )

parallel_chunk = "\n".join(
    ["lemma par_demo: True", "proof -"]
    + [slow_have(i) for i in range(1, N_HAVES + 1)]
    + ["show True by simp", "qed"]
)
print(parallel_chunk)
````

````python
await client.enter_theory(session_id, "Par_Demo", lease_id=lease_id)
await client.execute_command(session_id, "theory Par_Demo imports Complex_Main begin", lease_id=lease_id)
````

````python
import time

t0 = time.perf_counter()
res = await client.execute_command(session_id, parallel_chunk, timeout=120.0, lease_id=lease_id)
parallel_wall = time.perf_counter() - t0
print(f"one chunk, {N_HAVES} parallel haves: {parallel_wall:.1f}s "
      f"(sequential would be ~{N_HAVES * SLEEP_S}s of sleep alone)")
res
````

````python
# Sequential baseline: same steps, one command per request.
# Each step waits for checking to finish, so the sleeps cannot overlap.
t0 = time.perf_counter()
await client.execute_command(session_id, "lemma seq_demo: True", timeout=120.0, lease_id=lease_id)
await client.execute_command(session_id, "proof -", timeout=120.0, lease_id=lease_id)
for i in range(1, N_HAVES + 1):
    await client.execute_command(session_id, slow_have(i), timeout=120.0, lease_id=lease_id)
await client.execute_command(session_id, "show True by simp", timeout=120.0, lease_id=lease_id)
await client.execute_command(session_id, "qed", timeout=120.0, lease_id=lease_id)
sequential_wall = time.perf_counter() - t0
print(f"step-by-step: {sequential_wall:.1f}s | one chunk: {parallel_wall:.1f}s "
      f"| speedup x{sequential_wall / parallel_wall:.1f}")
````

````python
await client.execute_command(session_id, "end", lease_id=lease_id)
await client.release_session(session_id, lease_id=lease_id)
````

To see the knobs bite, edit `.env`, restart the server, and rerun the chunk cell (the pool is rebuilt on restart, so new sessions pick up the options):

| `.env` setting | expected chunk wall time |
|---|---|
| `ISABELLE_PARALLEL_PROOFS=2`, `ISABELLE_SESSION_THREADS=4` | ≈ `SLEEP_S` (all 4 haves in flight) |
| `ISABELLE_PARALLEL_PROOFS=2`, `ISABELLE_SESSION_THREADS=2` | ≈ `2 × SLEEP_S` (two at a time) |
| `ISABELLE_PARALLEL_PROOFS=0` | ≈ `N_HAVES × SLEEP_S` (no forking — matches the step-by-step baseline) |

### Recorded results (2026-06-10)

Environment: rebuilt `isabellegym-isabelle-gym` image (Isabelle2025-2, linux/amd64 under qemu on Apple Silicon, `mem_limit: 12g`), `N_HAVES = 4`, `SLEEP_S = 2`. The first two rows are the cell outputs above; the third row reran the identical chunk after restarting the server with `ISABELLE_PARALLEL_PROOFS=0` (script: `claude-work/run-parallelism-demo/run_noparallel.py`), then the server was restarted again with the default parallel settings.

| server config | run | wall time | server `execution_time` |
|---|---|---|---|
| `parallel_proofs=2`, `threads=4` | one chunk, 4 parallel `have`s | **6.9 s** | 6.66 s |
| `parallel_proofs=2`, `threads=4` | step-by-step (one command per request) | **26.6 s** | — |
| `parallel_proofs=0` (forking off) | one chunk, same 4 `have`s | **24.9 s** | 24.72 s |

Takeaways:

- With forking on, the chunk runs **×3.8** faster than the step-by-step baseline (26.6 s → 6.9 s) — the speedup matches `N_HAVES = 4`, i.e. all four `have` bodies were checked concurrently.
- With `parallel_proofs=0` the *same chunk* degrades to 24.9 s ≈ `N_HAVES × SLEEP_S` + overhead — indistinguishable from the step-by-step baseline, confirming the speedup comes from PIDE proof forking and not from chunked submission itself.
- The parallel chunk does not hit the theoretical ~`SLEEP_S` floor (2 s) because the amd64 image runs under qemu emulation on this host, which inflates per-step checking overhead (visible also in the 26.6 s baseline vs ~8 s of pure sleep).

## Big-chunk verification with `verify_chunk`

The sledgehammer section above advanced the `amc12b_2020_p22` proof **one command per call** — many round-trips, and any single line could time out on its own with no indication of *which* line hung.

`verify_chunk` submits the **whole proof body in one call**. It is checked with per-session parallelism on, bounded by a **single** wall budget, and returns a **per-command status report** in source order — each command is `ok`, `failed`, `running`, or `unprocessed`. On a timeout the report is *partial* and names the still-`running` line (the likely loop) instead of surfacing an opaque timeout.

Below we verify the **exact same proof** from the sledgehammer section as a single chunk.

````python
# verify_chunk needs the theory already begun, so acquire a session, enter the
# theory, and process the header as a normal step first.
chunk_payload = await client.acquire_session(["Complex_Main"], "HOL")
chunk_sid = chunk_payload["session_id"]
chunk_lease = chunk_payload["lease_id"]
await client.enter_theory(chunk_sid, "amc12b_2020_p22", lease_id=chunk_lease)
await client.execute_command(
    chunk_sid,
    'theory amc12b_2020_p22 imports Complex_Main "HOL-Library.Code_Target_Numeral" begin',
    lease_id=chunk_lease,
)
````

````python
# The same theorem..qed proof from the sledgehammer section, as ONE chunk.
proof_chunk = r"""theorem amc12b_2020_p22: fixes t :: real shows "((2 powr t - 3 * t) * t) / (4 powr t) ≤ 1 / 12"
proof -
have hpow: "(2 powr t)^2 = 4 powr t"
by (metis powr_mult numeral_Bit0_eq_double power2_eq_square)
have pos: "(0::real) < 4 powr t" by simp
have amgm: "2 * (2 powr t) * (6 * t) ≤ (2 powr t)^2 + (6 * t)^2"
by (rule sum_squares_bound)
have rearr: "12 * ((2 powr t - 3 * t) * t) = 2 * (2 powr t) * (6 * t) - (6 * t)^2"
by (simp add: power2_eq_square algebra_simps)
from amgm rearr hpow have key: "12 * ((2 powr t - 3 * t) * t) ≤ 4 powr t"
by linarith
from key pos show ?thesis
by (simp add: field_simps)
qed
"""

# Single call, single wall budget; per-command status comes back in source order.
report = await client.verify_chunk(chunk_sid, proof_chunk, timeout=180.0, lease_id=chunk_lease)

# Built-in pretty printer in the async client — no custom loop needed.
client.print_chunk_report(report)
report
````

````python
# Finish the theory and return the session to the pool.
await client.execute_command(chunk_sid, 'end', timeout=120.0, lease_id=chunk_lease)
await client.close_session(chunk_sid, lease_id=chunk_lease)
````
